# 리뷰데이터 없는 데이터 전처리(요약 리뷰랑 합치기 위해)

```python
df_noreview = pd.read_excel('./압구정_데이터/no_review.xlsx')

df_noreview
```

```python
waiting_map = {
        "대기 시간 바로 입장": 0,
        "대기 시간 10분 이내": 10,
        "대기 시간 30분 이내": 30,
        "대기 시간 30분 이상": 30,
        "대기 시간 1시간 이상": 60,
        "대기 시간 2시간 이상": 120
    }
df_noreview['웨이팅'] = df_noreview['웨이팅'].str.strip().map(waiting_map)


df_noreview
```

```python
import pandas as pd

# 1. 웨이팅 상위 1등/2등 추출
waiting_count = df_noreview.groupby("가게명")["웨이팅"].value_counts().reset_index(name="개수")
waiting_count = waiting_count.sort_values(['가게명', '개수'], ascending=[True, False])

# 각 가게별 상위 1등/2등 웨이팅 값 추출
top_2_waiting = waiting_count.groupby('가게명').head(2).reset_index(drop=True)

# 1등과 2등 웨이팅을 피벗으로 변환
top_2_wide = top_2_waiting.assign(rank=top_2_waiting.groupby('가게명').cumcount() + 1)  # 순위 부여
top_2_wide = top_2_wide.pivot(index='가게명', columns='rank', values='웨이팅')  # 피벗으로 1등/2등 나눔
top_2_wide.columns = ['1등 웨이팅', '2등 웨이팅']  # 컬럼 이름 정리
top_2_wide.reset_index(inplace=True)

# 2. 동료 상위 1등/2등 추출
grouped = df_noreview.groupby(['가게명', '동료']).size().reset_index(name='빈도')
grouped['총빈도'] = grouped.groupby('가게명')['빈도'].transform('sum')
grouped['비율'] = (grouped['빈도'] / grouped['총빈도'])

grouped_sorted = grouped.sort_values(['가게명', '빈도'], ascending=[True, False])
top_2 = grouped_sorted.groupby('가게명').head(2)

# 1등과 2등 동료 추출
top_2_pivot = top_2.assign(rank=top_2.groupby('가게명').cumcount() + 1)
top_2_wide_dongryo = top_2_pivot.pivot(index='가게명', columns='rank', values=['동료', '비율'])

# 컬럼 이름 정리
top_2_wide_dongryo.columns = ['1등 동료', '2등 동료', '1등 비율', '2등 비율']
top_2_wide_dongryo.reset_index(inplace=True)

# 3. 가게별 고유 데이터 유지 및 병합
unique_data = df_noreview.drop_duplicates(subset=['가게명']).reset_index(drop=True)
unique_data = unique_data[['가게명', '요약주소', '주소', '가게분류']]  # 필요한 컬럼만 유지

# 4. 데이터 병합: 웨이팅, 동료 데이터 추가
final_result = pd.merge(unique_data, top_2_wide, on='가게명', how='left')
final_result = pd.merge(final_result, top_2_wide_dongryo, on='가게명', how='left')

# 결과 출력
display(final_result)
```

```python
# 가게명을 기준으로 병합
merged_df = final_result.merge(df[['가게명', '리뷰']], on='가게명', how='left')

display(merged_df)

merged_df.to_excel('최종데이터12.xlsx')
```
